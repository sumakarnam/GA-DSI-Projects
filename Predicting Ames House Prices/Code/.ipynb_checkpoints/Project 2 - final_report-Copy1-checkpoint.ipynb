## Background and Problem Statement

The goal of this analysis is to predict housing prices in Ames, Iowa using extensive [data](../datasets/train.csv) about various features of the house (79 features about the size, quality, amenities etc.). A detailed description of the dataset can be found in the [data description](http://jse.amstat.org/v19n3/decock/DataDocumentation.txt)

The house prices are predicted using a Linear Regression model and the model was tuned to minimize root mean square error on the available data.

Problem Statement: As the observed house prices have a log normal distribution (jump to visualizations of this distribution in the [EDA section](#Exploratory-Data-Analysis)), in addition to predicting the house prices for the 878 houses in the [test data](../datasets/test.csv), this analysis aims to determine if predicting log prices instead of prices gives a better fitting model and to quanify the lift in $R^{2}$ if this is indeed the case.

## Functions

This sections contains some plotting and scoring functions used in the notebook

```python
def numerical_to_categorical_and_plot(category, target, df):
    '''
bins numerical columns into 10 bins with equal # observations and 
plots mean of target column for each bin vs bin average

TODO: 
include a strategy to deal with missing values when binning the category column. 
add the average number of observations in each bin to the graphs
Didn't look up what pd.qcut does with missing values

'''
    
    df_copy = df.copy()
    df_copy[f'{category}_bins'] = pd.qcut(df_copy[category],10, duplicates = 'drop')
    result_df = df_copy.groupby(f'{category}_bins').agg({target:['mean', 'count']})
    result_df[f'{category}_bin_average']=[round(a.mid) for a in result_df.index]
    result_df.columns = result_df.columns.to_flat_index()
    result_df.columns = [x[0]  if x[1] == '' else x[0]+'_'+x[1] for x in result_df.columns]
    fig, ax = plt.subplots()
    sns.lineplot(x = f'{category}_bin_average', y = f'{target}_mean' , data = result_df,         
                    ax = ax, color = '#0b5394');   
    return result_df


def plot_bar_line(category, target, df, sort = False):
    
    
    '''
    creates the following two plots on a single figure with two y-axes 
    for category and target columns of a df
    1. left y-axis: bar graph mean of target per category
    2. right y-axes: line plot of # of observations in each value of category in df
    
    If sort = True, the bars are ordered in descending values of the mean of target

   '''
    
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    
    if sort:

        df1 = (
            df.groupby([category])[target].mean()
            .to_frame().reset_index().sort_values(by = target, ascending = False)
        )
    else:
        df1 = (
            df.groupby([category])[target].mean()
            .to_frame().reset_index()
        )
        
    
    df2 =(df[category].value_counts(sort = False)
          .rename_axis(category).to_frame('Counts').reset_index()
    )
 
    
    
    df = df1.merge(df2, how ='left', on = category)
        
    df[category] = df[category].astype(str)

    sns.barplot(x = category, y = target, data = df, ci = False, 
                ax = ax1, color = '#0b5394')
    sns.lineplot(x = category , y = 'Counts', data = df, 
                 ax = ax2, color = 'red',marker = 'o'  );

    ax1.set_xlabel(category, color = 'black')
    ax1.set_ylabel(f'{target}_mean',color = '#0b5394')
    ax2.set_ylabel('# Observations',color = 'red');
    

def rmse(predictions, targets):
    '''
    returns the root mean square error 
    '''
    return np.sqrt(((predictions - targets) ** 2).mean())

def r2_adj(sklearn_linearmodel,X,y):
    ''' 
    parameters: 
    sklearn_linearmodel : instance of LinearRegression from sklearn
    X,y : features and true values as Dataframes
    
    returns adjust R^2 value
    '''
    sklearn_linearmodel.fit(X,y)
    R2 = sklearn_linearmodel.score(X,y)
    n = X.shape[0] #number of observations
    p = X.shape[1] #number of variables = number of coefficients in the model
    adj_R2 = 1 - ((1-R2)*((n-1)/(n-p-1)))        
    return adj_R2



def metrics(sklearn_linearmodel,X,y):
    sklearn_linearmodel.fit(X,y)
    y_pred = sklearn_linearmodel.predict(X)
    print(f'R2 : {r2_score(y,y_pred)}')
    print(f'R2_adj : {r2_adj(sklearn_linearmodel,X,y)}')
    print(f'MSE : {mean_squared_error(y,y_pred)}')
    print(f'RMSE : { np.sqrt(mean_squared_error(y,y_pred))}')
    print(f'Explained variance : {explained_variance_score(y,y_pred)}')
    print(f'Maximum Error : {max_error(y,y_pred)}')
    
```

### Treating Missing Values


```python
null_counts = ames_df.isnull().sum() 

#list of column names with nulls
columns_with_nulls = null_counts[null_counts>0].index 

print('columns with null values and the number of null values are:')
null_counts[columns_with_nulls]

#dropping columns in which more than 40% of the observations are missing
cols_to_drop = [col for col in columns_with_nulls if null_counts[col] > 0.4*ames_df.shape[0]]
print(f'#columns being dropped: {len(cols_to_drop)}')

# the rest of the columns with missing values to be treated with some imputing strategy
cols_to_impute = [col for col in columns_with_nulls if null_counts[col] <= 0.4*ames_df.shape[0]]
print(f'#columns to impute: {len(cols_to_impute)}')
ames_df[cols_to_impute].describe()
```

The following steps were taken to treat missing values:
- 5 columns had more than 40\% of values missing. These columns have not been included in the analysis
- In the rest of the 21 columns with missing values, the missing values were replcaed with the mode of the column for categorical and discrete numerical columns and with the mean of the column for continuous numerical variables

```python
# creating lists of various kinds of features 

cat_columns = ames_df.select_dtypes(object).columns.to_list()

discrete_columns = ['MS SubClass', 'Overall Qual','Overall Cond', 'Year Built', 'Year Remod/Add', 
       'Bsmt Full Bath', 'Bsmt Half Bath', 'Full Bath', 'Half Bath',
       'Bedroom AbvGr', 'Kitchen AbvGr', 'TotRms AbvGrd', 'Fireplaces',
       'Garage Yr Blt', 'Garage Cars',  'Misc Val', 'Mo Sold', 'Yr Sold']
continuous_columns = ['Lot Frontage', 'Lot Area', 'Mas Vnr Area',
       'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF',
       '1st Flr SF', '2nd Flr SF', 'Low Qual Fin SF', 'Gr Liv Area','Garage Area', 'Wood Deck SF',
       'Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 'Screen Porch',
       'Pool Area']
ames_df[continuous_columns].describe(percentiles = [0.01,0.1,0.25,0.5, 0.6, 0.75,0.9,0.99])
ames_df[discrete_columns].describe(percentiles = [0.01,0.1,0.25,0.5, 0.6, 0.75,0.9,0.99])

# checking that the above feature lists are correct
assert(ames_df.shape[1] == len(cat_columns)+len(discrete_columns)+len(continuous_columns) + len(['Id', 'PID', 'SalePrice']))

# drop columns with more than 40% missing values
df_no_nulls = ames_df.drop(columns = cols_to_drop)


# imputing missing values in the rest of the columns with nulls

imputed_df = df_no_nulls.copy()

for col in cols_to_impute:
#replacing null values in categorical and discrete numerical columns with the mode of the column
    if col in cat_columns + discrete_columns:
        imputed_df[col] = imputed_df[col].fillna(imputed_df[col].mode()[0])
#replacing the null values in continuous numerical columns with mean of the column    
    else:
        imputed_df[col] = imputed_df[col].fillna(imputed_df[col].mean())
        
imputed_df.dtypes
imputed_df.isnull().sum()
```

## Feature Engineering

The columns are transformed as described in the EDA. A total of 27 columns are being used of which four columns are being encoded with the mean price based on category ('Overall Qual' , 'MS SubClass', 'Kitchen Qual','Neighborhood'). All the other categorical columns are being binarized except 'Lot Shape' which is one-hot encoded. 

```python
## Feature engineering done based on the EDA

#numerical features
#choosing 13/39(28 without any nulls) columns
num_columns_to_mean_encode = ['Overall Qual' , 'MS SubClass']
num_columns_to_binarize = ['Full Bath' , 'Half Bath', 'Fireplaces']
num_continuous_variables = ['Gr Liv Area', 'Year Built','Year Remod/Add', 'Garage Area', 
                            'TotRms AbvGrd','Open Porch SF','Wood Deck SF','Lot Area'] #to be used as-is



#categorical features
#choosing 14/42 columns (27 with no nulls) based on analysis above
cat_cols_to_binarize = ['MS Zoning', 
       'Lot Config', 'Condition 1',
       'Bldg Type', 'House Style', 'Roof Style', 
       'Exterior 2nd', 'Exter Qual', 
       'Foundation', 'Heating QC',
       'Sale Type']
cat_columns_to_mean_encode = ['Kitchen Qual','Neighborhood']
cat_columns_to_dummify = ['Lot Shape']

features = ['Overall Qual' , 'MS SubClass','Full Bath' , 'Half Bath', 
            'Fireplaces','Gr Liv Area', 'Year Built','Year Remod/Add', 'Garage Area',
           'TotRms AbvGrd','Open Porch SF','Wood Deck SF','Lot Area' , 'MS Zoning', 
       'Lot Config', 'Condition 1', 'Bldg Type', 'House Style', 'Roof Style', 
       'Exterior 2nd', 'Exter Qual', 'Foundation', 'Heating QC',
       'Sale Type', 'Kitchen Qual','Neighborhood', 'Lot Shape']      
len(features)
```

```python
#creating train-test-split before mean encoding variables
X = imputed_df[features + ['SalePrice']].copy()
y = imputed_df['SalePrice'].copy()

#train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 2410)
X_train.describe()
```

```python
#Featurizing function to predict SalePrice:

Mean_maps = {col : X_train.groupby([col])[['SalePrice']].mean() 
             for col in ['Overall Qual' , 'MS SubClass', 'Kitchen Qual','Neighborhood']}.copy()

cat_columns_to_binarize_dict = {
                    'MS Zoning': ['FV', 'RL'], 
                    'Lot Config':['CulDSac'], 
                    'Condition 1':['PosN', 'PosA', 'RRNn', 'Norm'],
                    'Bldg Type': ['Twnhse','1Fam'] ,
                    'House Style': ['2.5Fin','2.5Unf', '2Story'], 
                    'Roof Style' : ['Hip'], 
                    'Exterior 2nd': ['CmentBd','ImStucc','VinylSd'], 
                    'Exter Qual':['Ex','Gd'], 
                    'Foundation':['PConc'], 
                    'Heating QC': ['Ex'] ,
                    'Sale Type' : ['New']
                   }
#mean encoded columns:
cols_to_mean_encode = ['Overall Qual' , 'MS SubClass', 'Kitchen Qual','Neighborhood']

# Keeping the numerical columns as-is and encoding categorical columns

def featurize(X_df, Mean_maps, cat_columns_to_binarize_dict, cols_to_mean_encode):
    result = X_df.copy()
    
    #binarizing
    for col, array in cat_columns_to_binarize_dict.items():
        result.loc[:,col] = result.loc[:,col].map(lambda x : 1 if x in array else 0)
    #mean encoding
    for col in cols_to_mean_encode:
        result.loc[:,col]=result.loc[:,col].map(lambda x : Mean_maps[col].loc[x,'SalePrice'] 
                                      if x in Mean_maps[col].index else imputed_df['SalePrice'].mean())
    #one-hot encoding
    result = pd.get_dummies(result, columns=['Lot Shape'], drop_first=True)    
    return result
```

```python
#Featurizing function to predict log(price):

Mean_maps = {col : X_train.groupby([col])[['SalePrice']].mean() 
             for col in ['Overall Qual' , 'MS SubClass', 'Kitchen Qual','Neighborhood']}.copy()

cat_columns_to_binarize_dict = {
                    'MS Zoning': ['FV', 'RL'], 
                    'Lot Config':['CulDSac'], 
                    'Condition 1':['PosN', 'PosA', 'RRNn', 'Norm'],
                    'Bldg Type': ['Twnhse','1Fam'] ,
                    'House Style': ['2.5Fin','2.5Unf', '2Story'], 
                    'Roof Style' : ['Hip'], 
                    'Exterior 2nd': ['CmentBd','ImStucc','VinylSd'], 
                    'Exter Qual':['Ex','Gd'], 
                    'Foundation':['PConc'], 
                    'Heating QC': ['Ex'] ,
                    'Sale Type' : ['New']
                   }
#mean encoded columns:
cols_to_mean_encode = ['Overall Qual' , 'MS SubClass', 'Kitchen Qual','Neighborhood']

# Keeping the numerical columns as-is and encoding categorical columns

def featurize_logprice(X_df, Mean_maps, cat_columns_to_binarize_dict, cols_to_mean_encode):
    result = X_df.copy()
    for col, array in cat_columns_to_binarize_dict.items():
        result.loc[:,col] = result.loc[:,col].map(lambda x : 1 if x in array else 0)
    for col in cols_to_mean_encode:
        result.loc[:,col]=result.loc[:,col].map(lambda x : np.log(Mean_maps[col].loc[x,'SalePrice']) 
                                      if x in Mean_maps[col].index else np.log(imputed_df['SalePrice'].mean()))
    result = pd.get_dummies(result, columns=['Lot Shape'], drop_first=True)    
    return result
```

## Model Fitting, Cross Validation and Iterations


The above features are used to fit a Linear Regression model to predict home price and log(home price).

- The $R^{2}$ and RMSE for both training and test data are in the output cells below for every model iteration
- The model coefficients are also displayed for ever iteration in the output cells

 ### Another Model iteration

In this attempt, the impact of 'Neighborhood' on the home price is beign captured by finding the average price per squarefoot for every neighborhood and creating a feature which is area of the home multipled by the price per sqft in its neighborhood.

This improved the $R^{2}$ of the model marginally to 0.82 and the RMSE to 35073$


```python
#featurizing neighborhood by nbhd_pricepersqft*Gr Liv Area

#creating a dictionary of mean_pricepersqft by neighborhood


new_nbhd_encoder_df= imputed_df[['Neighborhood']]
new_nbhd_encoder_df['Pricepersqft'] = imputed_df.loc[:,'SalePrice']/imputed_df.loc[:,'Gr Liv Area']
nbhd_dict = {col: new_nbhd_encoder_df.groupby('Neighborhood')['Pricepersqft'].mean()[col] 
             for col in new_nbhd_encoder_df.groupby('Neighborhood')['Pricepersqft'].mean().index }


Mean_maps = {col : X_train.groupby([col])[['SalePrice']].mean() 
             for col in ['Overall Qual' , 'MS SubClass', 'Kitchen Qual']}.copy()

cat_columns_to_binarize_dict = {
                    'MS Zoning': ['FV', 'RL'], 
                    'Lot Config':['CulDSac'], 
                    'Condition 1':['PosN', 'PosA', 'RRNn', 'Norm'],
                    'Bldg Type': ['Twnhse','1Fam'] ,
                    'House Style': ['2.5Fin','2.5Unf', '2Story'], 
                    'Roof Style' : ['Hip'], 
                    'Exterior 2nd': ['CmentBd','ImStucc','VinylSd'], 
                    'Exter Qual':['Ex','Gd'], 
                    'Foundation':['PConc'], 
                    'Heating QC': ['Ex'] ,
                    'Sale Type' : ['New']
                   }
#mean encoded columns:
cols_to_mean_encode = ['Overall Qual' , 'MS SubClass', 'Kitchen Qual']



def featurize_nbhd(X_df, Mean_maps, cat_columns_to_binarize_dict, cols_to_mean_encode, nbhd_dict):
    result = X_df.copy()
    for col, array in cat_columns_to_binarize_dict.items():
        result.loc[:,col] = result.loc[:,col].map(lambda x : 1 if x in array else 0)
    for col in cols_to_mean_encode:
        result.loc[:,col]=result.loc[:,col].map(lambda x : Mean_maps[col].loc[x,'SalePrice'] 
                                      if x in Mean_maps[col].index else imputed_df['SalePrice'].mean())
    result.loc[:,'Neighborhood'] =  result.loc[:,'Neighborhood'].map(lambda x:nbhd_dict[x])
    result.loc[:,'Neighborhood'] = result.loc[:,'Neighborhood']*result.loc[:,'Gr Liv Area']
    result = pd.get_dummies(result, columns=['Lot Shape'], drop_first=True)    
    return result
```

```python
#Fitting and scoring the model with the 'Gr Liv Area * nbhd_ppsgft' variable instead of Neighborhood and Gr Liv Area
cols_to_mean_encode = ['Overall Qual' , 'MS SubClass', 'Kitchen Qual']
#Featurizing
X_train.columns
X_featurized_train = featurize_nbhd(X_train.drop(columns = ['SalePrice']), 
                                    Mean_maps, cat_columns_to_binarize_dict, cols_to_mean_encode, nbhd_dict)
X_featurized_test = featurize_nbhd(X_test.drop(columns = ['SalePrice']),
                                    Mean_maps, cat_columns_to_binarize_dict, cols_to_mean_encode, nbhd_dict)
#Scaling

ss = StandardScaler()
Z_train = ss.fit_transform(X_featurized_train.drop(columns = ['Gr Liv Area']))
Z_test = ss.transform(X_featurized_test.drop(columns = ['Gr Liv Area']))

#Linear regression with SalePrice
lr = LinearRegression()
cross_val_score(X= Z_train, y = y_train,estimator = lr,cv = 7)
print(f'cross validation score is {np.mean(cross_val_score(X= Z_train, y = y_train,estimator = lr,cv = 7))} ')

lr.fit(Z_train, y_train)


#evaluating metrics
print(f"Training R-Squared: {lr.score(Z_train, y_train)} ")
print(f"Testing R-Squared: {lr.score(Z_test, y_test)} ")

print(f"Training RMSE: {rmse(lr.predict(Z_train),y_train)} ")
print(f"Testing RMSE: {rmse(lr.predict(Z_test),y_test)} ")



#sanity check: comparing results with the stats model
lr_stats = sm.OLS(y_train, sm.add_constant(Z_train)).fit()
lr_stats.summary()

fig, ax = plt.subplots(figsize = (8,5))
sns.scatterplot(y = lr.predict(Z_test), x = y_test  , ax = ax,  alpha = 0.5, color = myblue)
plt.title('Predictions vs True Price for area*nbhd_price_per_sqft model', fontweight = 'bold')
plt.ylabel('Predictions')
plt.xlabel('True Price'); 
```

```python
#Featurizing function to predict SalePrice:

Mean_maps = {col : X_train.groupby([col])[['SalePrice']].mean() 
             for col in ['Overall Qual' , 'MS SubClass', 'Kitchen Qual','Neighborhood']}.copy()



cat_columns_to_binarize_dict = {
                    'MS Zoning': ['FV', 'RL'], 
                    'Lot Config':['CulDSac'], 
                    'Condition 1':['PosN', 'PosA', 'RRNn', 'Norm'],
                    'Bldg Type': ['Twnhse','1Fam'] ,
                    'House Style': ['2.5Fin','2.5Unf', '2Story'], 
                    'Roof Style' : ['Hip'], 
                    'Exterior 2nd': ['CmentBd','ImStucc','VinylSd'], 
                    'Exter Qual':['Ex','Gd'], 
                    'Foundation':['PConc'], 
                    'Heating QC': ['Ex'] ,
                    'Sale Type' : ['New']
                   }
#mean encoded columns:
cols_to_mean_encode = ['Overall Qual' , 'MS SubClass', 'Kitchen Qual','Neighborhood']

# Keeping the numerical columns as-is and encoding categorical columns

def featurize(X_df, Mean_maps, cat_columns_to_binarize_dict, cols_to_mean_encode):
    result = X_df.copy()
    
    #binarizing
    for col, array in cat_columns_to_binarize_dict.items():
        result.loc[:,col] = result.loc[:,col].map(lambda x : 1 if x in array else 0)
    #mean encoding
    for col in cols_to_mean_encode:
        result.loc[:,col]=result.loc[:,col].map(lambda x : Mean_maps[col].loc[x,'SalePrice'] 
                                      if x in Mean_maps[col].index else imputed_df['SalePrice'].mean())
    #one-hot encoding
    result = pd.get_dummies(result, columns=['Lot Shape'], drop_first=True)    
    return result
```
